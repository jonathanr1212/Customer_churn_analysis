# Overview

This project uses Machine Learning (ML) to make predictions on a binary classification target for a telecommunications company. The information came from a dataset provided on kaggle from SyriaTel. Models are used to make the best predictions to find out why customers have churned.(left the company)

# Business Problem

SyriaTel wants to reduce the amount of money lost by checking what are the factors that have made consumers churn. This analysis can help predict which customers have the possibility of churning and find a way to retain them for the future. 

# Data Preparation and EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, f1_score, 
    mean_squared_error, confusion_matrix, plot_confusion_matrix, roc_auc_score, classification_report
)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
```

```python
df = pd.read_csv('data/bigml_59c28831336c6604c800002a.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['churn'].value_counts()
```

```python
#Formatting target boolean values to integer values to be able to put through machine learning modeling
df_2 = df.copy()
```

```python
df_2['churn'] = df_2['churn'].astype('int64')
```

```python
#Checking Value counts to see if the dataset has a class imbalance. 
df_2.churn.value_counts()
```

```python
#Converting binary information into integers
df_2['international plan'], df_2['voice mail plan'] = (df_2['international plan'].map(dict(no= 0, yes= 1)), 
                                                      df_2['voice mail plan'].map(dict(no= 0, yes= 1)))
```

```python
#double checking to see if everything was encoded properly
print(df['international plan'].value_counts())
print(df['voice mail plan'].value_counts())
print(df_2['international plan'].value_counts())
print(df_2['voice mail plan'].value_counts())
```

```python
#dropping phone numbers and area codes
df_2.drop(['phone number', 'area code'], axis=1, inplace=True)
df_2.info()
```

```python
df_2.hist();
```

```python
#Cgecking counts for the categorical data
sns.countplot(y='state', data=df_2, color='tab:blue')
sns.set(rc={'figure.figsize':(30,40)}, font_scale= 3);
```

```python
#Checking churn by state
states = df_2.groupby('state')['churn'].value_counts(normalize=True)
states = pd.DataFrame(states)
states.columns = ['percent']
states = states.reset_index()
states
```

```python
#Checking churn rate by state
states_churn_rate = states.loc[states['churn'] == 1].sort_values("percent", ascending =False).reset_index().drop(
    "index", axis =1)
states_churn_rate
```

```python
#categorizing states churn rate into high at 20% and above, mid to 10%-19%, and low to 9% and lower.
high_churn, mid_churn, low_churn = (states_churn_rate.loc[states_churn_rate['percent'] >= .2], 
                                    
                                    states_churn_rate.loc[(states_churn_rate['percent'] < .2) & (
                                        states_churn_rate['percent'] >= .1)], 
                                    
                                    states_churn_rate.loc[states_churn_rate['percent'] < .1])

high_churn_states, mid_churn_states, low_churn_states  = (high_churn['state'],
                                                          mid_churn['state'],
                                                          low_churn['state'])

print("States with a high churn rate: ", list(high_churn_states))
print('\n----------------------------------------------------------------\n')
print("States with a medium churn rate: ", list(mid_churn_states))
print('\n-----------------------------------------------------------------\n')
print("States with a low churn rate: ", list(low_churn_states))
```

```python
num_list = df_2.select_dtypes(include=['int64', 'float64', 'int32'])
for col in num_list:
    fig, axes = plt.subplots(1, 1, figsize=(4,4))
    sns.boxplot(y = df_2['churn'].astype('category'), x =col , data=df_2, orient='h')
    sns.set(font_scale= 1.5)
    
```

```python
#checking for possible outlier removal
df_2_value_counts = df_2[['customer service calls', 'total intl calls', 'total eve charge', 'total day calls']]
for counts in (df_2_value_counts):
    print(df_2_value_counts[counts].value_counts(), "-------------------------")
```

### Model Preparation

```python
#setting x and y for model preparation
X = df_2.drop('churn', axis=1)
y = df_2['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
X_train_num = X_train.select_dtypes(include=['float64', 'int64', 'int32'])
X_train_cat = X_train.select_dtypes('object')

```

```python
num_pipeline = Pipeline(steps=[
    ('ss', StandardScaler())
])
                
cat_pipeline = Pipeline(steps=[
    ('ohe', OneHotEncoder(handle_unknown='ignore',
                         sparse=False))
])

transformer = ColumnTransformer(transformers=[
    ('numerical', num_pipeline, X_train_num.columns),
    ('categorical', cat_pipeline, X_train_cat.columns)])
```

# Modeling

```python
#Starting off with a dummy model to check how accurate the model would be if we just chose the most frequent.
dummy_pipe = Pipeline(steps=[
    ('transformer', transformer),
    ('dummy', DummyClassifier(strategy="most_frequent"))]) 

dummy_pipe.fit(X_train, y_train)
dummy_pipe.predict(X_train)
```

```python
cv_results = cross_val_score(dummy_pipe, X_train, y_train, cv=5)
cv_results.mean()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
plot_confusion_matrix(dummy_pipe, X_train, y_train, normalize='true', ax=ax)
plt.grid(False);
```
